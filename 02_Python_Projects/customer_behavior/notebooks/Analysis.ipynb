```python
import pandas as pd
df = pd.read_csv("C:\\Users\\abhay\\OneDrive\\Desktop\\Data-Analytics-Portfolio\\02_Python_Projects\\customer_behavior\\data\\Messy_Ecommerce_Data.csv")
```

## 🧱 STEP 1 — Load & Inspect Data

```python
print(df.shape)
print(df.describe())
print(df.info())
```

```python
print(df["city"].isnull().sum())
print(df["discount_percent"].isnull().sum())
print(df.duplicated().sum())
```

## 🧹 STEP 2 — Remove Duplicate Rows

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

## 🧠 STEP 3 — Fix Order Date

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed')
```

## 🧹 STEP 4 — Handle Missing Values

```python
df["discount_percent"] = df["discount_percent"].fillna(0)
df['city'] = df['city'].fillna("Unknown")
```

## 🧹 STEP 5 — Fix Text Columns

```python
df['product_name'] = df['product_name'].str.strip().str.lower()
df['product_category'] = df['product_category'].str.strip().str.lower()
df['city'] = df['city'].str.strip().str.lower()
```

## 🧹 STEP 6 — Fix Negative Quantity

```python
df = df[df['quantity'] > 0]
```

## 🧠 STEP 7 — Create New Column (Total_sales, Month)

```python
df["total_sales"] = df["quantity"]*df["unit_price"]*(1-df["discount_percent"]/100)
```

```python
df["Month"] = df["order_date"].dt.to_period("M")
```

## 🎯 Analysis Task 1 — Monthly Revenue

```python
df.groupby("Month")["total_sales"].sum().sort_values(ascending=False)
```

## 🎯 Analysis Task 2 — Top customer

```python
df.groupby('customer_id')['total_sales'].sum().sort_values(ascending=False).head(10)
```

```python
# top 10 customer by sales
print(df.groupby("customer_id")["total_sales"].sum().sort_values(ascending=False).head(10))

# top 5 city by sales
print(df.groupby("city")["total_sales"].sum().sort_values(ascending=False).head(5))

# top 5 product category by sales
print(df.groupby("product_category")["total_sales"].sum().sort_values(ascending=False).head(5))
```

## Key Insights
- June 2023 generated the highest revenue (~90K).
- Clothing is the top-performing product category.
- Customer CUST22 is the highest revenue-generating customer.
- A large portion of revenue comes from customers with missing city information, indicating a potential data quality issue.
