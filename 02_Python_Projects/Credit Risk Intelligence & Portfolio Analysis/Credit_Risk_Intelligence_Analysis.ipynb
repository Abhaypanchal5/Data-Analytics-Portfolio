```python
import pandas as pd
```

## 🟢 Task 01 — Data Integrity Audit

```python
df = pd.read_csv("Retail_Bank_Credit_Risk_500_Customers.csv")

print("Dataset Shape:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Customer_IDs:")
print(df.duplicated(subset="Customer_ID").sum())
```

## Task 02 — Descriptive Statistical Intelligence

```python
print(df.describe())
print(df["Annual_Income"].mean())
print(df["Credit_Score"].mean())
print(df["Loan_Amount"].mean())
```

## 🟢 Task 03 — Outlier Detection

```python
Q1 = df["Loan_Amount"].quantile(0.25)
Q3 = df["Loan_Amount"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Loan_Amount"] < lower_bound) | 
              (df["Loan_Amount"] > upper_bound)]

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of Outliers:", outliers.shape[0])
```

## 🔥 Task 04 — Portfolio Risk Profiling

```python
df["Default_Flag"] = df["Default_Status"].map({"Yes": 1, "No": 0})

print("Overall Default Rate:",
      df["Default_Flag"].mean() * 100, "%")

print("\nDefault Rate by Gender:")
print(df.groupby("Gender")["Default_Flag"].mean() * 100)

print("\nDefault Rate by Employment Type:")
print(df.groupby("Employment_Type")["Default_Flag"].mean() * 100)

print("\nDefault Rate by City Tier:")
print(df.groupby("City_Tier")["Default_Flag"].mean() * 100)
```

```python
city_analysis = df.groupby("City_Tier").agg({
    "Default_Flag": ["mean", "sum", "count"],
    "Loan_Amount": "sum"
})

print(city_analysis)
```

## 🔥 Task 06 — Income vs Default Risk

```python
income_analysis = df.groupby(pd.qcut(df["Annual_Income"], 4)).agg({
    "Default_Flag": ["mean", "count"],
    "Loan_Amount": "mean"
})

print(income_analysis)

# “Default probability declines significantly with income. 
# However, loan amounts remain consistent across income bands, indicating elevated leverage risk among lower-income borrowers.”
```
