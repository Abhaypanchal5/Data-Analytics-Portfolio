```python
import csv
import pandas as pd 
```

```python
data = pd.read_csv("Bank_loan.csv")
```

## 🔍 Task 01 — Data Integrity & Quality Assessment

```python
# Missing value report = No null values
data.info()
```

```python
# no duplicate is there in data
def any_dupe(Col_name):
    return data.duplicated(subset=Col_name).sum()

any_dupe("Loan_ID")
```

## 💰 Task 02 — Financial Exposure Engineering

```python
data["Loan_Exposure"] = data["Loan_Amount"]  #(For now they are equal. Later we’ll improve this.)

print(f"The Total Loan Exposure is {data['Loan_Exposure'].sum():,.2f}")
print(f"The Average Loan Exposure is {data['Loan_Exposure'].mean():,.2f}")
print(f"The Minimum Loan Exposure is {data['Loan_Exposure'].min():,.2f}")
print(f"The Maximum Loan Exposure is {data['Loan_Exposure'].max():,.2f}")
```

The bank currently holds a total loan exposure of ₹2,010,000.
The average loan size of ₹201,000 suggests moderate lending per customer.
The loan distribution ranges from ₹90,000 to ₹350,000, indicating variation in customer borrowing capacity.
Higher loan amounts may contribute disproportionately to financial risk if default occurs.

## Task-03-Default-Risk-Segmentation

```python
data["Loan_Status"] = (data["Default_Status"] == "Yes").astype(int)

exposure_summary = data.groupby("Loan_Status")["Loan_Exposure"].sum()
print(exposure_summary)

default_rate = data["Loan_Status"].mean() * 100
print(f"Default Rate: {default_rate:.2f}%")
```

## 📊 Task 04 — Risk Concentration Analysis by Income Segment

```python
data["Income_Bucket"] = pd.cut(
    data["Income"],
    bins=[0, 30000, 60000, float("inf")],
    labels=["Low Income", "Medium Income", "High Income"]
)

print(data.groupby("Income_Bucket")["Loan_Exposure"].sum())
print(data.groupby("Income_Bucket")["Loan_Exposure"].count())
default_rate_income = data.groupby("Income_Bucket")["Loan_Status"].mean() * 100
print(default_rate_income)
```

The Low Income segment shows a 100% default rate, indicating extremely high credit risk within this group.

The Medium Income segment has a moderate default rate of 25%, suggesting manageable but present risk.

The High Income segment has 0% default rate, indicating strong repayment capacity.

Although High Income customers hold the largest exposure (₹900,000), the financial risk is concentrated within the Low Income segment due to its complete default occurrence.

## 🔥 Task 05 — Credit Score Risk Profiling

```python
data["Credit_Category"] = pd.cut(
    data["Credit_Score"],
    bins=[0, 600, 700, float("inf")],
    labels=["Poor", "Fair", "Good"]
)

print("Total Loan Exposure by Credit Category:")
print(data.groupby("Credit_Category")["Loan_Exposure"].sum())

default_rate_credit = data.groupby("Credit_Category")["Loan_Status"].mean() * 100

print("\nDefault Rate by Credit Category:")
print(default_rate_credit)
```

Credit score appears to be a strong predictor of default behavior.

Both Poor and Fair credit categories show a 100% default rate,
while the Good credit category has 0% defaults.

This suggests that credit score segmentation is highly effective
for risk assessment within this dataset.

Although the bank’s highest exposure (₹1,500,000) lies in the Good
credit category, the associated risk remains minimal due to zero defaults.

This indicates strong portfolio quality driven by high-credit borrowers.

## 🔥 Task 06 — Risk-Weighted Exposure Calculation

```python
data["Risk_Weighted_Exposure"] = data["Loan_Exposure"] * data["Loan_Status"]

total_risk_exposure = data["Risk_Weighted_Exposure"].sum()
portfolio_total = data["Loan_Exposure"].sum()

risk_percentage = (total_risk_exposure / portfolio_total) * 100

print(f"Total Exposure at Risk: ₹{total_risk_exposure:,.2f}")
print(f"Portfolio Percentage at Risk: {risk_percentage:.2f}%")
```

The total risk-weighted exposure represents the amount currently tied to defaulted loans.

The portfolio percentage at risk indicates how much of the bank’s lending book is exposed to potential loss.

Higher percentages signal concentration of financial risk and may require policy adjustments.

## 📊 Task 07 — Risk Driver Comparison Analysis

```python
income_default = data.groupby("Income_Bucket")["Loan_Status"].mean() * 100

income_highest = income_default.max()
income_lowest = income_default.min()
income_spread = income_highest - income_lowest

print("Income Default Rates:")
print(income_default)
print(f"\nIncome Spread: {income_spread:.2f}%")

credit_default = data.groupby("Credit_Category")["Loan_Status"].mean() * 100

credit_highest = credit_default.max()
credit_lowest = credit_default.min()
credit_spread = credit_highest - credit_lowest

print("\nCredit Default Rates:")
print(credit_default)
print(f"\nCredit Spread: {credit_spread:.2f}%")
```
