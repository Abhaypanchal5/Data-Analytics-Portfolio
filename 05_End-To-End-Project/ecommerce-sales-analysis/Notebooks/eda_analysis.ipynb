## 🧠 Step 2 — Exploratory Data Analysis (EDA)

```python
import pandas as pd

df = pd.read_csv("C:\\Users\\abhay\\OneDrive\\Desktop\\Data-Analytics-Portfolio\\05_End-To-End-Project\\ecommerce-sales-analysis\\Data\\Cleaned_data.csv",encoding="ISO-8859-1")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df.head(10)
```

## Monthly Revenue

```python
df.groupby(df["InvoiceDate"].dt.month)["Revenue"].sum()
```

## Top Products

```python
df.groupby("Description")["Revenue"].sum().sort_values(ascending=False).head(10)
```

## Top Countries

```python
df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)
```

## Top Customers

```python
df.groupby("CustomerID")["Revenue"].sum().sort_values(ascending=False).head(10)
```
